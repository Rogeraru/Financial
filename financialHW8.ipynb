```python
X1 ='Total current assets'
X2 ='Total current liabilities'
X3 ='Total assets'
X4 ='Total retained earnings'
X5 ='Earnings before interest and tax'
X6 ='Total equity'
X7 ='Total liabilities'
X8 ='Total operating revenue'
TSMC ='2330 TSMC'
MTK ='2454 MTK'
```

```python
A_TSMC = (df1.loc[TSMC,X1] - df1.loc[TSMC,X2])/df1.loc[TSMC, X3]


B_TSMC = df1.loc[TSMC,X4]/df1.loc[TSMC, X3]


C_TSMC = df2.loc[TSMC,X5]/df1.loc[TSMC, X3]


D_TSMC = df1.loc[TSMC,X6]/df1.loc[TSMC, X7] #調整分子項定義，採用股東權益帳面價值


E_TSMC = df2.loc[TSMC,X8]/df1.loc[TSMC, X3]


Z_score_TSMC = 1.2*A_TSMC + 1.4*B_TSMC + 3.3*C_TSMC + 0.6*D_TSMC +1.0*E_TSMC


Z_score_TSMC, A_TSMC, B_TSMC, C_TSMC, D_TSMC, E_TSMC
```

```python

A_MTK = (df1.loc[MTK,X1] - df1.loc[MTK,X2])/df1.loc[MTK, X3]


B_MTK = df1.loc[MTK,X4]/df1.loc[MTK, X3]


C_MTK = df2.loc[MTK,X5]/df1.loc[MTK, X3]


D_MTK = df1.loc[MTK,X6]/df1.loc[MTK, X7]


E_MTK = df2.loc[MTK,X8]/df1.loc[MTK, X3]


Z_score_MTK = 1.2*A_MTK + 1.4*B_MTK + 3.3*C_MTK + 0.6*D_MTK +1.0*E_MTK


Z_score_MTK, A_MTK, B_MTK, C_MTK, D_MTK, E_MTK
```
